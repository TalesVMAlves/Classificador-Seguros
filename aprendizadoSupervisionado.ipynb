```python
import pandas as pd

df = pd.read_csv('artefatos/insurance_data_preprocessado.csv')
df.head()
```

```python
df['cluster'].value_counts()
```

```python
from sklearn.model_selection import train_test_split

X = df.drop(['claim', 'cluster', 'binned_claim'], axis=1) ## Evitar o vazamento de informação
y = df.cluster ## Vamos utilizar o cluster como label

X_train, X_val, y_train, y_val = train_test_split(
    X, 
    y, 
    test_size=0.8, 
    random_state=1408,
    stratify=df['cluster'] 
) # 80% Treino 20% Validacao
```

```python
from pipelineTreino import train_and_evaluate_models
df_results, creports = train_and_evaluate_models(X_train, y_train, X_val, y_val)
```

```python
df_results.sort_values(by='MCC', ascending=False)
```

```python
from pipelineTreino import perform_bayesian_optimization
from sklearn.svm import LinearSVC
from skopt.space import Real, Categorical, Integer
linear_svc_model = LinearSVC(random_state=1408, max_iter=10000)

# Busca por hiperparâmetros
search_parameters = {
    'C': Real(1e-4, 1e+1, prior='log-uniform'), 
    # 'penalty': Categorical(['l1', 'l2']),  # L1 não é suportado para hinge loss ou dual True
    'loss': Categorical(['hinge', 'squared_hinge']),  
    'tol': Real(1e-6, 1e-2, prior='log-uniform'),  
    'fit_intercept': Categorical([True, False]),  
    'max_iter': Integer(100, 5000),
}
# Perform Bayesian optimization
best_model_assunto = perform_bayesian_optimization(linear_svc_model, search_parameters, X_train, y_train, X_val, y_val)
```
